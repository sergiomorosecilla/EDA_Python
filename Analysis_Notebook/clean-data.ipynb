```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns

pd.set_option('display.max_columns', None)
```

1.Procedo a cargar los datos del archivo CSV y realizo algunas modificaciones en las columnas. Despues de ell mostramos resumen de la info.

```python
#Cargo el archivo csv
df_bank_additional = pd.read_csv('./bank-additional.csv')
```

```python
#Me aseguro que las columnas tengan nombres en minúsculas
df_bank_additional.columns = df_bank_additional.columns.str.lower()
# Elimino columna 'unnamed: 0'que es innecesaria
df_bank_additional.drop(columns ='unnamed: 0' , axis=1, inplace=True)
#Renombro la columna 'y' a 'subscribed' para mayor claridad
df_bank_additional.rename(columns={'y': 'subscribed'}, inplace=True) 
#Renombro la columna 'id_' por 'id' para poder hacer merge con el otro dataset
df_bank_additional.rename(columns={'id_': 'id'}, inplace=True)
# Verifico que no haya duplicados
print(df_bank_additional.duplicated().sum())                                                                                 
#Obtenemos la info sobre la cantidad de datos (columnas y filas) y tipos de variables 
df_bank_additional.info() 
#Mostramos un resumen de las primeras filas y columnas del archivo
df_bank_additional.head()
```

```python
#Pido un resumen de las metricas mas importantes de los datos, para tener una visual inicial y tenerlo a mano
round(df_bank_additional.describe(), 2).T
```

2. Realizo lo mismo para el archivo XLSX.

```python
#Procedo con los mismos pasos para el archivo xlsx
# Leer todas las hojas del Excel
excel_path = 'customer-details.xlsx'
all_sheets = pd.read_excel(excel_path, sheet_name=None)

# Unir todas las hojas en un solo DataFrame
df_customer_details = pd.concat(all_sheets.values(), ignore_index=True)

# Verificar duplicados en IDs
print(df_customer_details['ID'].duplicated().sum())
```

```python
#Me aseguro que las columnas tengan nombres en minúsculas
df_customer_details.columns = df_customer_details.columns.str.lower()
# Elimino columna 'unnamed: 0'que es innecesaria
df_customer_details.drop(columns ='unnamed: 0' , axis=1, inplace=True)
#Renombro la columna 'ID' por 'id' para poder hacer merge con el otro dataset
df_bank_additional.rename(columns={'ID': 'id'}, inplace=True)                                                                                       
#Obtenemos la info sobre la cantidad de datos (columnas y filas) y tipos de variables 
df_bank_additional.info() 
#Mostramos un resumen de las primeras filas y columnas del archivo
df_bank_additional.head()
```

```python
round(df_customer_details.describe(), 2).T
```

3. Uno los dos datasets a través de sus respectivas columnas 'id' que es la referencia de las clientes y puede servir como clave única. Acabaré de realizar limpieza final de datos

```python
# Realizar inner join (solo registros comunes)
df_merged = pd.merge(df_bank_additional, df_customer_details, on='id', how='inner')
# Tamaños
print("Tamaño df_bank:", df_bank_additional.shape)
print("Tamaño df_customer:", df_customer_details.shape)
print("Tamaño df_merged (inner):", df_merged.shape)

df_merged.info()
df_merged.head()
df_merged.describe

# Guardar el DataFrame combinado en un nuevo archivo CSV
df_merged.to_csv('bank_campaign_merged.csv', index=False)
```
